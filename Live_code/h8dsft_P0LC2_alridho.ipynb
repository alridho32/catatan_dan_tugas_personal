# SQL

## Problem 1 - SQL

1. Kota apa saja yang masuk ke dalam cakupan wilayah distribusi?

```python
df = client.query('''

SELECT DISTINCT store_city AS Kota_Distribusi
FROM `ftds-hacktiv8-project.phase0_ftds_037_rmt.beverage_sales`

''').to_dataframe()

df
```

Kota yang masuk ke dalam cakupan wilayah distribusi adalah `Clarinda, Muscatine, Mason City, West Des Moines, Clinton, dan Cedar Falls`. Total terdapat `6 Kota`

2. Berapa rentang harga minimum dan maksimum dari unit minumam yang dijual?

```python
df = client.query('''

SELECT MIN(unit_retail_price) AS minimum_price, MAX(unit_retail_price) AS maximum_price
FROM `ftds-hacktiv8-project.phase0_ftds_037_rmt.beverage_sales`

''').to_dataframe()

df
```

Rentang harga minuman yang di jual adalah dari `0.98` hingga `54.0`

3. Bandingkan total penjualan dan jumlah transaksi dari masing-masing kota pada tahun 2022?

```python
df = client.query('''

SELECT store_city AS city, SUM(total_price) AS total_penjualan_2022, SUM(unit_sold) AS jumlah_transaksi_2022, COUNT(invoice_id) AS total_transaksi,
FROM `ftds-hacktiv8-project.phase0_ftds_037_rmt.beverage_sales`
WHERE EXTRACT(YEAR FROM date) IN (2022)
GROUP BY store_city;

''').to_dataframe()

df
```

Berikut adalah perbandingan total penjualan dan jumlah transaksi dari masing-masing kota pada tahun 2022

4. Produk apa yang paling banyak diminati pada tahun 2022?

```python
df = client.query('''

SELECT product_name, SUM(unit_sold) AS jumlah_unit_terjual_2022
FROM `ftds-hacktiv8-project.phase0_ftds_037_rmt.beverage_sales`
WHERE EXTRACT(YEAR FROM date) IN (2022)
GROUP BY product_name
ORDER BY jumlah_unit_terjual_2022 DESC
LIMIT 1;
''').to_dataframe()

df
```

Produk yang paling banyak diminati pada tahun 2022 adalah `CitrusBurst Breeze` dengan unit terjual sebanyak `7415`

## Problem 2 PANDAS

```python
import pandas as pd
df = client.query('''SELECT * FROM `ftds-hacktiv8-project.phase0_ftds_037_rmt.beverage_sales`''').to_dataframe()

df
```

```python
df.info()
```

1. buat kolom baru, menunjukkan representasi musim dengan ketentuan berikut:

```python
df['date'] = pd.to_datetime(df['date'])

season = []
for bulan in df['date'].dt.month:
  if bulan in [3,4,5]:
    season.append('Spring')
  elif bulan in [6,7,8]:
    season.append('Summer')
  elif bulan in [9,10,11]:
    season.append('Autumn')
  else:
    season.append('Winter')

df['Season'] = season
df
```

Berikut merupakan tabel beverage_sales yang telah di tambahkan column `Season` pada bagian paling kanan

2. Berapa total barang yang terjual kepada toko Peak Provisions sepanjang tahun 2022? (Gunakan pandas query untuk menjawab soal no. 2)

```python
# Karena diperintahkan menggunakan pandas query maka cara yang dilakukan adalah seperti ini, tapi saya juga sudah menaruh cara non-querynya di bawah cara query, cara tersebut yang saya biasa gunakan
df['date'] = pd.to_datetime(df['date'])

peak_provisions = df.query('store_name == "Peak Provisions" and date.dt.year == 2022')
# peak_provisions = df[(df['store_name']=='Peak Provisions') & (df['date'].dt.year == 2022)] -----> Jika ingin menggunakan cara non-query

total_barang_terjual = peak_provisions['unit_sold'].sum()

total_barang_terjual
```

Total barang yang terjual kepada toko Peak Provisions sepanjang tahun 2022 adalah `4592 Barang`
