##### Sorting (DATA MANIPULATION)

```python
df.sort_values(by='id') # Sort berdasarkan id --> defaultnya ascending
# parameter sort_valuesnya banyak, bisa lihat di https://pandas.pydata.org/pandas-docs/version/2.0/reference/api/pandas.DataFrame.sort_values.html
df.sort_values(by=['id','conference']) # Untuk multiple sort bisa menggunakan list
df.sort_values(by=['id','conference'], ascending=[True, False]) # Untuk set multiple asc/desc

df_sort = df.sort_values(by=['id','conference'], ascending=[True, False])
```

```python
# Membuat column baru
df['new_column'] = df.id + 100
df
```

```python
# Untuk membuat 1 row baru
new_row = {
  "division": "one",
  "conference": "meja bundar",
  "school": "binus",
  "roster_url":"www.google.com"
  }

df._append(new_row, ignore_index=True)
```

```python
# Remove column
df.drop('new_column', axis=1, inplace=True) # HATI-HATI Merubah df di awal!!!!!!!!!
```

```python
# Keyword : del
del df['roster_url']
```

```python
# Boolean indexing / filtering
df = df[df.school_name != "Memphis"]
```

```python
df.agg({
  "id":"sum",
  "new_column": "mean"
})
```

```python
# Mencari mean dari masing-masing conference
df.groupby('conference')['new_column'].mean()
```
